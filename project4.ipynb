```python
# Q5
# select _id, code, creator, product_name, brands from products where manufacturing_places = 'Austin, TX' and stores = 'Whole Foods';

!mongo open_food --quiet --eval 'db.products.find({"manufacturing_places": "Austin, TX", "stores": "Whole Foods"}, {"code": 1, "creator": 1, "product_name": 1, "brands": 1}).pretty()'
```

```python
# Q6
# select _id, product_name, brands from products where brands in ("m&m's", "mars", "Mars", "oreo", "starburst") order by product_name limit 5;

!mongo open_food --quiet --eval 'db.products.find({"brands": {$in: ["m&m\u0027s", "mars", "Mars", "oreo", "starburst"]}}, {"product_name":1, "brands":1}).sort({"product_name":1}).limit(5).pretty()'
```

```python
# Q7
# select _id, product_name, countries, ingredients.text,
# ingredients.rank, nutriments.carbohydrates_serving
# from products
# where nutriments.carbohydrates_serving between 18 and 25 order by ingredients.rank DESC
# limit 5;

!mongo open_food --quiet --eval 'db.products.find({ "nutriments.carbohydrates_serving" : { $gt :  18, $lt : 25}}, {"product_name":1, "countries":1, "ingredients.text":1, "ingredients.rank":1, "nutriments.carbohydrates_serving":1}).sort({"ingredients.rank":-1}).limit(5).pretty()'
```
